#1] Imports

```python
#Imports
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import mean_absolute_error
```

# 2] Pre-Processing

```python
# Reading in relevant data
data = pd.read_csv("WildfireData.csv", na_values="NaN")
data = data[['fire_size', 'remoteness', 'discovery_month', 'Vegetation', 'putout_time']]
```

```python
# Drop NA
data = data.dropna()
```

```python
# One Hot Encodings
non_dummy_cols = ['fire_size', 'remoteness','putout_time'] 
dummy_cols = list(set(data.columns) - set(non_dummy_cols))
data = pd.get_dummies(data, columns=dummy_cols)
```

```python
# Putout conversion
data['putout_time'] = data['putout_time'].str.split(' ').str[0]
```

#3] Models

```python
#Split train/test
from sklearn.model_selection import train_test_split
X = data.drop('putout_time', axis=1) 
y = data.putout_time.astype('float')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
#Construct Model
model = Sequential()
model.add(Dense(20, input_dim=(21), activation='relu', kernel_initializer='he_normal'))
model.add(Dense(10, activation='relu', kernel_initializer='he_normal'))
model.add(Dense(1, activation='linear'))
```

```python
# Compile Model
model.compile(loss='mse', optimizer='adam')
```

```python
# Fit Model
model.fit(X_train, y_train, epochs=200, batch_size=128, verbose=1)
```

```python
# Show MAE error
yhat = model.predict(X_test)
error = mean_absolute_error(y_test, yhat)
print('MAE: %.3f' % error)
```

```python
# Save Model
model.save('nn.h5')
```
